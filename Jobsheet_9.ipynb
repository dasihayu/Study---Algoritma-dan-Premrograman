**Eksperimen Dengan Modul `math`**

```python
import math

def luas_lingkaran(r):
    return math.pi * r ** 2

print("Luas lingkaran dengan r = 7 :", luas_lingkaran(7))
print("Luas lingkaran dengan r = 10:", luas_lingkaran(10))
```

**Eksperimen Dengan Modul `random`**



```python
import random

for i in range(5):
    angka = random.randint(1, 100)
    print("Angka acak ke-", i+1, ":", angka)
```

**Modul Buatan Sendiri**

```python
%%writefile konversi.py

def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi

print("25°C =", konversi.celcius_ke_fahrenheit(25), "°F")
print("77°F =", konversi.fahrenheit_ke_celcius(77), "°C")
```
