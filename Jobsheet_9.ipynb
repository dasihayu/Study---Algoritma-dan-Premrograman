**Modul Buatan Sendiri**

```python
%%writefile konversi.py

def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi

print("25°C =", konversi.celcius_ke_fahrenheit(25), "°F")
print("77°F =", konversi.fahrenheit_ke_celcius(77), "°C")
```

**Studi Kasus Mini - Program Kuis**

```python
%%writefile laporan.py

def tampilkan_ringkasan(jumlah_soal, skor):
    print("\n===== Ringkasan Kuis =====")
    print(f"Jumlah Soal: {jumlah_soal}")
    print(f"Jawaban Benar: {skor}")
    print(f"Nilai Akhir: {(skor / jumlah_soal) * 100:.2f}")
    print("==========================")
```

```python
import random
import laporan

jumlah_soal = 5
skor = 0

print("=== Kuis Penjumlahan Sederhana ===")

for i in range(1, jumlah_soal + 1):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    jawaban_benar = a + b

    print(f"Soal {i}: {a} + {b} = ?")
    jawaban_user = int(input("Jawaban kamu: "))

    if jawaban_user == jawaban_benar:
        print("✅ Benar!\n")
        skor += 1
    else:
        print(f"❌ Salah! Jawaban yang benar adalah {jawaban_benar}\n")

laporan.tampilkan_ringkasan(jumlah_soal, skor)
```
