**Praktikum 04 – Fungsi dengan args dan kwargs**

```python
def jumlahkan(*args):
    return sum(args)
def biodata (**kwargs):
    for key, value in kwargs.items():
        print(f"{key}  : {value}")

print("Total : ", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi", Umur="20", Prodi="TRK")
```

**Praktikum 7 - Decorator**

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")
sapa()
```

**Fungsi Dengan `args`**

**Fungsi Dengan `kwargs`**

```python
def info_barang(**data):
    for kunci, nilai in data.items():
        print(f"{kunci}: {nilai}", end=" ")
    print()

info_barang(Nama="Buku", Harga=15000, Stok=30)
info_barang(Nama="Pensil", Harga=3000, Stok=100, Warna="Kuning")
info_barang(Nama="Laptop", Harga=8500000)
```

**Fungsi `lambda`**

```python
import math

kuadrat = lambda x: x ** 2
keliling_lingkaran = lambda r: 2 * math.pi * r

print("Kuadrat dari 5 =", kuadrat(5))
print("Keliling lingkaran dengan jari-jari 7 =", keliling_lingkaran(7))
```

**Fungsi `decorator`**

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo! Selamat datang di Python.")

halo()
```

**Studi Kasus Nyata – Program Nilai**

```python
def input_nilai():
    nilai = []
    while True:
        teks = input("Masukkan nilai mahasiswa (tekan Enter jika selesai): ")
        if teks == "":
            break
        try:
            nilai.append(float(teks))
        except ValueError:
            print("Input tidak valid! Masukkan angka saja.")
    return nilai

def hitung_rata(nilai):
    if len(nilai) == 0:
        return 0
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

print("=== Program Nilai Mahasiswa ===")
daftar_nilai = input_nilai()

if daftar_nilai:
    rata = hitung_rata(daftar_nilai)
    print(f"Rata-rata nilai: {rata:.2f}")
    print("Status:", status_kelulusan(rata))
else:
    print("Tidak ada nilai yang dimasukkan.")

try:
    tugas = float(input("Masukkan nilai tugas: "))
    ujian = float(input("Masukkan nilai ujian: "))
    akhir = nilai_akhir(tugas, ujian)
    print(f"Nilai akhir berdasarkan bobot: {akhir:.2f}")
except ValueError:
    print("Input nilai tugas/ujian tidak valid!")
```
