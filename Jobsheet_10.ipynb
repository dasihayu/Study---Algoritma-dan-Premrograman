**Praktikum 1 - Bubble Sort**

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range (0, n-i-1):
      if data[j] > data[j+1]:
        data[j], data[j+1] = data[j+1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 21, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

**Praktikum 2 - Selection Sort**

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i+1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

**Praktikum 3 - Insertion Sort**

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j+1] = data[j]
      j -= 1
    data[j+1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

**Praktikum 4 - Studi Kasus Nilai Mahasiswa**

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range (0, n-i-1):
      if data[j][1] > data[j+1][1]:
        data[j], data[j+1] = data[j+1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

**Praktikum 6 - Sorting Data Barang**

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

**Eksperimen Bubble Sort**

```python
def bubble_sort_desc(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Sebelum diurutkan :", angka)
print("Sesudah Bubble Sort (desc):", bubble_sort_desc(angka.copy()))
```

**Eksperimen Selection Sort**

```python
def selection_sort_largest_to_end(arr):
    n = len(arr)
    for i in range(n):
        max_index = 0
        for j in range(0, n - i):
            if arr[j] > arr[max_index]:
                max_index = j

        arr[max_index], arr[n - i - 1] = arr[n - i - 1], arr[max_index]

    return arr


angka = [64, 25, 12, 22, 11, 90, 34]
print("Data sebelum diurutkan :", angka)
print("Data setelah Selection Sort:", selection_sort_largest_to_end(angka))
```

**Eksperimen Insertion Sort**

```python
def insertion_sort_alfabet(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j].lower() > key.lower():
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data


nama_mahasiswa = ["Cici", "Andi", "Doni", "Budi", "Eka"]
print("Sebelum sorting :", nama_mahasiswa)
print("Setelah sorting  :", insertion_sort_alfabet(nama_mahasiswa))
```

```python
%%writefile selection_sort_count.py

def selection_sort_count(data):
    n = len(data)
    count = 0
    for i in range(n):
        max_index = 0
        for j in range(0, n - i):
            count += 1
            if data[j] > data[max_index]:
                max_index = j
        data[max_index], data[n - i - 1] = data[n - i - 1], data[max_index]
    return count
```

File insertion_sort_count.py

```python
%%writefile insertion_sort_count.py

def insertion_sort_count(data):
    n = len(data)
    count = 0
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            count += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return count
```
